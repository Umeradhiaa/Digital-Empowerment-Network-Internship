```python
fig, ax = plt.subplots(1,2,figsize=(40, 5))
ax[0].pie(df['sex'].value_counts(), labels = ['Male', 'Female'], autopct='%1.1f%%', startangle=90)
ax[0].set_title('Gender Distribution')
sns.countplot(x = 'age', data = df, ax = ax[1]).set_title('Age Distribution')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=90, ha='right')
```

```python
fig, ax = plt.subplots(1,3,figsize=(40, 5))
sns.histplot(x = 'trestbps', data = df, ax = ax[0], kde = True).set_title('trestbps Distribution')
sns.histplot(x = 'thalach', data = df, ax = ax[1], kde = True).set_title('thalach Distribution')
sns.histplot(x = 'oldpeak', data = df, ax = ax[2], kde = True).set_title('oldpeak Distribution')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns = ['target']), df['target'], test_size = 0.2, random_state = 0)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier
# Create Decision Tree object
dtc = DecisionTreeClassifier(random_state=0, max_depth= 12, min_samples_leaf=2, min_samples_split=2, class_weight='balanced')
```

```python
# Training the model
dtc.fit(X_train, y_train)
```

```python
# Training accuracy
dtc.score(X_train, y_train)*100
```

```python
# Predicting the test set results
dtc_pred = dtc.predict(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
#Training the model
lr.fit(X_train, y_train)
```

```python
#Training accuracy
lr.score(X_train, y_train)*100
```

```python
#Predicting the test set results
lr_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
fig, ax  = plt.subplots(1,2, figsize = (20,5))
sns.heatmap(confusion_matrix(y_test, dtc_pred), annot = True, cmap = 'coolwarm', ax = ax[0]).set_title('Decision Tree')
sns.heatmap(confusion_matrix(y_test, lr_pred), annot = True, cmap = 'coolwarm', ax = ax[1]).set_title('Logistic Regression')
```

```python
print('Decision Tree')
print('Accuracy Score: ', accuracy_score(y_test, dtc_pred)*100)
print('Precision Score: ', precision_score(y_test, dtc_pred)*100)
print('Recall Score: ', recall_score(y_test, dtc_pred)*100)
print('F1 Score: ', f1_score(y_test, dtc_pred)*100)
```

```python
print('Logistic Regression')
print('Accuracy Score: ', accuracy_score(y_test, lr_pred)*100)
print('Precision Score: ', precision_score(y_test, lr_pred)*100)
print('Recall Score: ', recall_score(y_test, lr_pred)*100)
print('F1 Score: ', f1_score(y_test, lr_pred)*100)
```
